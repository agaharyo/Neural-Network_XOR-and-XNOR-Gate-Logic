# XOR Gate

```python
import numpy as np
import pandas as pd
```

```python
def ANDNOT(x):
    w = [2, -1]
    z = np.dot(x,w)
    return binstep(z, 2)

def binstep(z, th):
    return 1 if z>= th else 0
```

```python
def XOR(x):
    z = [ANDNOT(x), ANDNOT(x[::-1])]
    return OR(z)
def OR(y):
    w = [1,1]
    z = np.dot(y,w)
    return binstep(z, 1)
def binstep(y,th):
    return 1 if y>=th else 0
```

```python
print(XOR([1,1]))
print(XOR([1,0]))
print(XOR([0,1]))
print(XOR([0,0]))
```

```python
df = pd.DataFrame([[1,1,XOR([1,1])],
                   [1,0, XOR([1,0])],
                   [0,1, XOR([0,1])],
                   [0,0, XOR([0,0])]], columns=['x1','x2','output XOR'])
df
```
